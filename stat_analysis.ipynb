```python
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
```

<h2> EXPOSED - SPREADER SIMULATIONS </h2>

```python
rumor_free = 'outputs/JSONS/ES/RUMORFREE/completeG_ES_RumorFree_'
truth_free = 'outputs/JSONS/ES/TRUTHFREE/completeG_ES_TruthFree_'
truthrumor_co = 'outputs/JSONS/ES/TRUTHRUMORCO/completeG_ES_TruthRumorCoexist_'
truthrumor_free = 'outputs/JSONS/ES/TRUTHRUMORFREE/completeG_ES_TruthRumorFree_'
```

```python
#load jsons
data_rf = []
for i in range(1,6):
    with open(rumor_free+str(i)+'.json') as f:
        data_rf.append(json.load(f))
        
data_tf = []
for i in range(1,6):
    with open(truth_free+str(i)+'.json') as f:
        data_tf.append(json.load(f))
        
data_trf = []
for i in range(1,6):
    with open(truthrumor_co+str(i)+'.json') as f:
        data_trf.append(json.load(f))
        
data_trc = []
for i in range(1,6):
    with open(truthrumor_free+str(i)+'.json') as f:
        data_trc.append(json.load(f))
```

## Get Averages Per Compartment

```python
I_rf = np.average(np.array([data_rf[i][0]['I'] for i in range(5)]), axis=0)
Er_rf = np.average(np.array([data_rf[i][0]['Er'] for i in range(5)]), axis=0)
Et_rf = np.average(np.array([data_rf[i][0]['Et'] for i in range(5)]), axis=0)
Sr_rf = np.average(np.array([data_rf[i][0]['Sr'] for i in range(5)]), axis=0)
St_rf = np.average(np.array([data_rf[i][0]['St'] for i in range(5)]), axis=0)
R_rf = np.average(np.array([data_rf[i][0]['R'] for i in range(5)]), axis=0)
```

```python
I_tf = np.average(np.array([data_tf[i][0]['I'] for i in range(5)]), axis=0)
Er_tf = np.average(np.array([data_tf[i][0]['Er'] for i in range(5)]), axis=0)
Et_tf = np.average(np.array([data_tf[i][0]['Et'] for i in range(5)]), axis=0)
Sr_tf = np.average(np.array([data_tf[i][0]['Sr'] for i in range(5)]), axis=0)
St_tf = np.average(np.array([data_tf[i][0]['St'] for i in range(5)]), axis=0)
R_tf = np.average(np.array([data_tf[i][0]['R'] for i in range(5)]), axis=0)
```

```python
I_trf = np.average(np.array([data_trf[i][0]['I'] for i in range(5)]), axis=0)
Er_trf = np.average(np.array([data_trf[i][0]['Er'] for i in range(5)]), axis=0)
Et_trf = np.average(np.array([data_trf[i][0]['Et'] for i in range(5)]), axis=0)
Sr_trf = np.average(np.array([data_trf[i][0]['Sr'] for i in range(5)]), axis=0)
St_trf = np.average(np.array([data_trf[i][0]['St'] for i in range(5)]), axis=0)
R_trf = np.average(np.array([data_trf[i][0]['R'] for i in range(5)]), axis=0)
```

```python
I_trc = np.average(np.array([data_trc[i][0]['I'] for i in range(5)]), axis=0)
Er_trc = np.average(np.array([data_trc[i][0]['Er'] for i in range(5)]), axis=0)
Et_trc = np.average(np.array([data_trc[i][0]['Et'] for i in range(5)]), axis=0)
Sr_trc = np.average(np.array([data_trc[i][0]['Sr'] for i in range(5)]), axis=0)
St_trc = np.average(np.array([data_trc[i][0]['St'] for i in range(5)]), axis=0)
R_trc = np.average(np.array([data_trc[i][0]['R'] for i in range(5)]), axis=0)
```

<h4> Plot </h4>

```python
xvalues = range(50)
fig_size= [15,12]
plt.rcParams.update({'font.size': 15, 
                     "figure.figsize": fig_size,
                     'axes.titlesize' : 23,
                     'axes.titlepad':10})
fig, axs = plt.subplots(2, 2)
# trf ||rf
# _________
#
# tf ||trc
axs[0,0].plot(xvalues, I_trf, color='deepskyblue', linestyle=':', label="I")
axs[0,0].plot(xvalues, Er_trf, color='b', linewidth=0.9, marker='o', label="E rumor")
axs[0,0].plot(xvalues, Et_trf, color='b', linestyle='--', label="E truth")
axs[0,0].plot(xvalues, Sr_trf, color='r', linewidth=0.4, marker='*', label="S rumor")
axs[0,0].plot(xvalues, St_trf, color='r', linewidth=0.7,linestyle='-', label="S truth")
axs[0,0].plot(xvalues, R_trf, color='coral', linestyle='-.', label="R")
axs[0,0].title.set_text("Truth and Rumor Free")


axs[1,0].plot(xvalues, I_tf, color='deepskyblue', linestyle=':', label="I")
axs[1,0].plot(xvalues, Er_tf, color='b', linewidth=0.9, marker='o', label="E rumor")
axs[1,0].plot(xvalues, Et_tf, color='b', linestyle='--', label="E truth")
axs[1,0].plot(xvalues, Sr_tf, color='r', linewidth=0.4, marker='*', label="S rumor")
axs[1,0].plot(xvalues, St_tf, color='r', linewidth=0.7,linestyle='-', label="S truth")
axs[1,0].plot(xvalues, R_tf, color='coral', linestyle='-.', label="R")
axs[1,0].title.set_text("Truth Free")

axs[0,1].plot(xvalues, I_rf, color='deepskyblue', linestyle=':', label="I")
axs[0,1].plot(xvalues, Er_rf, color='b', linewidth=0.9, marker='o', label="E rumor")
axs[0,1].plot(xvalues, Et_rf, color='b', linestyle='--', label="E truth")
axs[0,1].plot(xvalues, Sr_rf, color='r', linewidth=0.4, marker='*', label="S rumor")
axs[0,1].plot(xvalues, St_rf, color='r', linewidth=0.7,linestyle='-', label="S truth")
axs[0,1].plot(xvalues, R_rf, color='coral', linestyle='-.', label="R")
axs[0,1].title.set_text("Rumor Free")
axs[0,1].legend(loc=1)

axs[1,1].plot(xvalues, I_trc, color='deepskyblue', linestyle=':', label="I")
axs[1,1].plot(xvalues, Er_trc, color='b', linewidth=0.9, marker='o', label="E rumor")
axs[1,1].plot(xvalues, Et_trc, color='b', linestyle='--', label="E truth")
axs[1,1].plot(xvalues, Sr_trc, color='r', linewidth=0.4, marker='*', label="S rumor")
axs[1,1].plot(xvalues, St_trc, color='r', linewidth=0.7,linestyle='-', label="S truth")
axs[1,1].plot(xvalues, R_trc, color='coral', linestyle='-.', label="R")
axs[1,1].title.set_text("Truth and Rumor Coexist")

for i in range(0,2):
    for j in range(0,2):
        axs[i,j].margins(x=0.01,y=0.01)

fig.suptitle('Exposed-Spreader', fontsize=25)
plt.setp(axs,yticks=np.arange(0,1,0.2))
fig.tight_layout(rect=[0,0, 1, 0.96])
# fig.tight_layout()
plt.savefig('ES_completegraph',dpi=350)
# plt.show()
```
